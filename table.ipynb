Фигура 2. Гистограмма распределения балла тревожности (оформить красиво, подписать)

```python
anxiety = pd.read_csv('datasets/anxiety.csv', sep=';')
```

```python
import seaborn as sns

# Установим стиль для графика
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

sns_plot = sns.distplot(anxiety['stai_state_anxiety'].dropna(), color='violet')

plt.title('Distribution of the degree of anxiety')
plt.xlabel('Assessment of anxiety level')
plt.ylabel('Frequency')
sns.set(rc={'figure.figsize':(6,6)})
plt.show()
```

```python
sns.histplot(
    anxiety['stai_state_anxiety'], kde=True,
    stat="density", kde_kws=dict(cut=1), color='violet'
)

sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

sns.set(rc={'figure.figsize':(8,6)})

plt.title('Distribution of the degree of anxiety')
plt.xlabel('Assessment of anxiety level')
plt.ylabel('Frequency')
```

```python
# Создайте новый столбец 'rs8192678_dominant' с информацией о доминантных генотипах
anxiety['rs8192678_dominant'] = anxiety['rs8192678'].apply(lambda genotype: 'CC' if genotype == 'CC' else 'CT/TT')

# Проверьте, что новый столбец создан верно
print(anxiety['rs8192678_dominant'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Установите стиль для графика
sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

# Создайте подграфик
plt.figure(figsize=(3, 5))

# Определите цвета для "ящиков с усами" и точек
boxplot_colors = {'CC': 'gray', 'CT/TT': 'green'}
stripplot_colors = {'CC': 'red', 'CT/TT': 'blue'}

# "Ящики с усами" с цветами для обоих генотипов (CC и CT/TT)
sns.boxplot(x='rs8192678_dominant', y='stai_state_anxiety', data=anxiety, showfliers=False, palette=boxplot_colors)

# Точки с разными цветами для обоих генотипов (CC и CT/TT)
sns.stripplot(x='rs8192678_dominant', y='stai_state_anxiety', data=anxiety, palette=stripplot_colors, jitter=True, dodge=True)


plt.xlabel('Genotype (rs8192678)')
plt.ylabel('State Anxiety Score')

plt.show()
```
