```python
anxiety = pd.read_csv('anxiety.csv', sep=';')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

anxiety['stai_state_anxiety'] = pd.to_numeric(anxiety['stai_state_anxiety'], errors='coerce')

sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
sns.set(rc={'figure.figsize':(6,6)})

# Create the distribution plot after removing NaN values
sns_plot = sns.distplot(anxiety['stai_state_anxiety'].dropna(), color='black')

plt.title('Distribution of the degree of anxiety')
plt.xlabel('Assessment of anxiety level')
plt.ylabel('Frequency')

plt.show()
```

```python
sns.histplot(
    anxiety['stai_state_anxiety'], kde=True,
    stat="density", kde_kws=dict(cut=1), color='black'
)

sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

sns.set(rc={'figure.figsize':(8,6)})

plt.title('Distribution of the degree of anxiety')
plt.xlabel('Assessment of anxiety level')
plt.ylabel('Frequency')
```

```python

anxiety['rs8192678_dominant'] = anxiety['rs8192678'].apply(lambda genotype: 'CC' if genotype == 'CC' else 'CT/TT')
print(anxiety['rs8192678_dominant'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})
plt.figure(figsize=(3, 5))

boxplot_colors = {'CC': 'gray', 'CT/TT': 'green'}
stripplot_colors = {'CC': 'red', 'CT/TT': 'blue'}

sns.boxplot(x='rs8192678_dominant', y='stai_state_anxiety', data=anxiety, showfliers=False, palette=boxplot_colors)

sns.stripplot(x='rs8192678_dominant', y='stai_state_anxiety', data=anxiety, palette=stripplot_colors, jitter=True, dodge=True)


plt.xlabel('Genotype (rs8192678)')
plt.ylabel('State Anxiety Score')

plt.show()
```

```python
anxiety['rs17602729_dominant'] = anxiety['rs17602729'].apply(lambda genotype: 'GG' if genotype == 'GG' else 'GA')
print(anxiety['rs17602729_dominant'].value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("darkgrid", {"grid.color": ".6", "grid.linestyle": ":"})

plt.figure(figsize=(3, 5))

boxplot_colors = {'GG': 'gray', 'GA': 'green'}
stripplot_colors = {'GG': 'red', 'GA': 'blue'}

anxiety['rs17602729_dominant'].drop

sns.boxplot(x='rs17602729_dominant', y='stai_state_anxiety', data=anxiety, showfliers=False, palette=boxplot_colors)
sns.stripplot(x='rs17602729_dominant', y='stai_state_anxiety', data=anxiety, palette=stripplot_colors, jitter=True, dodge=True)

plt.xlabel('Genotype (rs17602729)')
plt.ylabel('State Anxiety Score')

plt.show()
```
